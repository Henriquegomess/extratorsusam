```python
import tabula
import pandas as pd
tabula.convert_into("OC_LEITOS_UND_20200601.pdf","OC_LEITOS_UND.csv",output_format = "csv",pages = "1",stream=True)
dfLeitosUnd = pd.read_csv('OC_LEITOS_UND.csv',keep_default_na=False)
#renomear as colunas
dfLeitosUnd.columns = ['unidade/indicador','total_sala_vermelha','sala_vermelha_covid','sala_vermelha_naocovid','total_leitos_clinicos','leitos_clinicos_covid','total_leitos_clinicos_naocovid','oc_leitos_clinicos_naocovid','livres_leitos_clinicos_naocovid','total_uti','leitos_uti_covid','leitos_uti_naocovid','ocupacao_sala_vermelha','ocupacao_leitos_clinicos','ocupacao_uti']

#deletar as 8 colunas com valores nulos e nomes das colunas
dfLeitosUnd = dfLeitosUnd.drop([0,1,2,3,4,5,6,7])
```

```python
#algumas colunas contem 3 dados, vamos separar cada valor em uma nova coluna
def separarColunas(newCol1, newCol2, newCol3, oldCol, df):
    df[[newCol1,newCol2,newCol3]] = df[oldCol].str.split(' ', n=3, expand=True)
    #deletar a coluna antiga
    df = df.drop(oldCol, axis=1)
    return df
```

```python
#separar as colunas com 3 valores
#(sala vermelha covid)
dfLeitosUnd = separarColunas('total_sala_vermelha_covid','oc_sala_vermelha_covid','livres_sala_vermelha_covid','sala_vermelha_covid',dfLeitosUnd)

#(sala vermelha nao covid)
dfLeitosUnd = separarColunas('total_sala_vermelha_naocovid','oc_sala_vermelha_naocovid','livres_sala_vermelha_naocovid','sala_vermelha_naocovid',dfLeitosUnd)

#(leitos clinicos covid)
dfLeitosUnd = separarColunas('total_leitos_clinicos_covid','oc_leitos_clinicos_covid','livres_leitos_clinicos_covid','leitos_clinicos_covid',dfLeitosUnd)

#(leito uti covid)
dfLeitosUnd = separarColunas('total_leitos_uti_covid','oc_leitos_uti_covid','livres_leitos_uti_covid','leitos_uti_covid',dfLeitosUnd)

#(leitos uti nao covid)
dfLeitosUnd = separarColunas('total_leitos_uti_naocovid','oc_leitos_uti_naocovid','livres_leitos_uti_naocovid','leitos_uti_naocovid',dfLeitosUnd)

dfLeitosUnd
```
